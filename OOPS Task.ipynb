```python
class Book:
  def __init__(self,name,title,Auther_name,price):


    name = input('Enter book name : ')
    title = input('Enter title : ')
    Auther_name = input('Enter Auther_name : ')
    price = int(input('Enter the price of book : '))


    self.name = name
    self.title = title
    self.Auther_name = Auther_name
    self.price = price


  def mybook(self):
    print('Book name is %s title is %s Auther_name is %s price is %d'%(self.name,self.title,self.Auther_name,self.price))

n = int(input('Enter how many books you want to add : '))

for i in range(n):
  a = Book('abc','short story','tgeh','100')

  a.mybook()



```
